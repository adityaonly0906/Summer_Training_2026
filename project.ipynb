```python
import pandas as pd

df = pd.read_excel("samplesuperstore.csv.xlsx")
df.dropna(inplace=True)
df.drop_duplicates(inplace=True)

print("Total Revenue =", df["Sales"].sum())
print("\nBest Selling Products",df.groupby("Product Name")["Quantity"].sum().nlargest(10))
print("\nTop Customers",df.groupby("Customer Name")["Sales"].sum().nlargest(10))

print("\nCategory Revenue",df.groupby("Category")["Sales"].sum())

df["Month"] = pd.to_datetime(df["Order Date"]).dt.month_name()
print("\nMonthly Sales Trend",df.groupby("Month")["Sales"].sum())
```
