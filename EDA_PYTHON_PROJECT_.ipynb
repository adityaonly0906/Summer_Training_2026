```python
#Load Dataset
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
sns.set_style("whitegrid")
```

```python

price_df = pd.read_csv("used_car_price_prediction_1M.csv")
price_df
```

```python
#Dataset Overview
# SHAPE
price_df.shape
```

```python
# COLUMNS
price_df.columns
```

```python
#DATA TYPE
price_df.info()
```

```python
#STATICTICAL SUMMARY
price_df.describe()
```

```python
#DATA CLEANING
#MISSING VALUES
price_df.isnull().sum()
```

```python
#DUPLICATE VALUES
price_df.duplicated().sum()
```

```python
#REMOVING DUPLICATES
price_df = price_df.drop_duplicates()
```

```python
# RELATION BETWEEN PRICE vs MILEAGE(with different transmission)
sns.scatterplot(data=price_df.sample(5000), x="Mileage_kmpl", y="Price",hue='Transmission')
plt.title("Mileage vs Price")
plt.show()
```

The scatter plot shows the relationship between fuel efficiency (Mileage_kmpl) and vehicle price. The color coding by transmission type reveals pricing differences between manual and automatic vehicles. Automatic cars generally appear in higher price ranges, while mileage has a moderate influence on price.

```python
# RELATION BETWEEN PRICE AND YEAR
sns.scatterplot(
    data=price_df.sample(5000),
    x="Year",
    y="Price",
    hue="Transmission",
);
```

The above scatter plot shows a positive relationship between vehicle year and price. Newer vehicles generally have higher market values than older vehicles. The transmission type also influences pricing, with certain transmission categories appearing more frequently in higher price ranges.

```python
# WHICH CAR BRAND APPEARS MOST FREQUENTLY
top_brands = price_df["Brand"].value_counts().head(10)

plt.figure(figsize=(10,5))
top_brands.plot(kind="bar")
plt.title("Top 10 Car Brands")
plt.xlabel("Brand")
plt.ylabel("Count")
plt.show()
```

The above bar chart displays the top 10 most frequently occurring car brands in the dataset. Brands with the highest counts have a larger presence in the used car market, indicating greater popularity or availability compared to other brands.

```python
#HOW ARE CAR PRICES DISTRIBUTED?
plt.figure(figsize=(10,5))

plt.hist(price_df["Price"], bins=30)

plt.xlabel("Price")
plt.ylabel("Frequency")
plt.title("Distribution of Car Prices")

plt.show()
```

The histogram shows the distribution of car prices in the dataset. Most vehicles are concentrated within a specific price range, while fewer vehicles exist at very high price levels. This indicates that the used car market is dominated by low- to mid-priced vehicles.

```python
#DOES HORSEPOWER AFFECT PRICE?
sns.scatterplot(data=price_df.sample(5000), x="Horsepower", y="Price")
plt.title("Horsepower vs Price")
plt.show()
```

The scatter plot suggests a positive relationship between horsepower and vehicle price. As horsepower increases, the price of the vehicle tends to increase as well. This indicates that engine performance is an important factor influencing car value.

```python
#HOW DOES REGISTRATION AFFECT PRICE?
avg_price = price_df.groupby("Registration_Age")["Price"].mean()

plt.figure(figsize=(12,6))

plt.plot(avg_price.index, avg_price.values, 'o-r')

plt.title("Average Price vs Registration Age")
plt.xlabel("Registration Age")
plt.ylabel("Average Price")

plt.show()
```

As registration age increases, the average price decreases.
